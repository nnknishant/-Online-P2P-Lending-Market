```python
x = [4,6,8,9,2,7]
y = [2,10,2,15,6,9]
plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Numbers")
c = ['red', 'blue', 'magenta', 'yellow','pink', 'green']

plt.scatter(x,y, color = c)
plt.show()
```

```python
x = [4,6,8,9,2,7]
y = [2,10,2,15,6,9]
plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Numbers")
c = ['red', 'blue', 'magenta', 'yellow','pink', 'green']

plt.scatter(x,y, color = c, s = 150)
plt.show()
```

```python
x = [10,20,30,40,50]
y = ['english', 'math', 'hindi', 'sanskrit', 'physics']
plt.title('Pie Chart')
c = ['yellow', 'pink', 'orange', 'purple', 'magenta']
plt.xlabel('length')
plt.ylabel('bredth')

plt.pie(x, labels= y, colors=c)
plt.legend()
plt.show()
```
